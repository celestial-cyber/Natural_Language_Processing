```python
import re

def whitespace_tokenize(text):
    return text.split()

def simple_regex_tokenize(text):
    return re.findall(r"\b\w+\b", text)

def regex_advanced_tokenization(text):
    pattern = r"[A-Za-z]+(?:'[A-Za-z]+)?|\d+(?:\.\d+)?"
    return re.findall(pattern, text)
    
def nltk_tokenization(text):
    try:
        import nltk
        from nltk.tokenize import word_tokenize
        nltk.download("punkt", quiet=True)
        nltk.download("punkt_tab", quiet=True)
        return word_tokenize(text)
    except Exception as e:
        return ["NLTK not available:", str(e)]

def spacy_tokenization(text):
    try:
        import spacy
        try:
            nlp = spacy.load("en_core_web_sm")
        except:
            import os
            os.system("python -m spacy download en_core_web_sm")
            nlp = spacy.load("en_core_web_sm")
        doc = nlp(text)
        return [token.text for token in doc]
    except Exception as e:
        return ["Spacy not available:", str(e)]

# Test text
test_text = "Hello, world! It's 2025. Can't wait!"

# Print statements for every function
print("1. Whitespace Tokenization:", whitespace_tokenize(test_text))
print("2. Simple Regex Tokenization:", simple_regex_tokenize(test_text))
print("3. Advanced Regex Tokenization:", regex_advanced_tokenization(test_text))
print("4. NLTK Tokenization:", nltk_tokenization(test_text))
print("5. spaCy Tokenization:", spacy_tokenization(test_text))
```
